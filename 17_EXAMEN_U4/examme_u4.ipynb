Problema: Clasificación de Clientes según su Perfil Financiero
Descripción del Problema
Una institución financiera desea clasificar a sus clientes en tres categorías basándose en su perfil financiero y de comportamiento:

Riesgo Bajo: Clientes que cumplen con todos los pagos a tiempo labels tienen ingresos estables.
Riesgo Medio: Clientes con retrasos esporádicos en los pagos o ingresos variables.
Riesgo Alto: Clientes con historial de impagos o ingresos inestables.
Características de Entrada:

Historial de pagos: Porcentaje de pagos realizados a tiempo (normalizado entre 0 labels 1).
Ingresos mensuales: Ingresos promedio del cliente (normalizado entre 0 labels 1).
Relación deuda-ingreso: Proporción entre la deuda total y los ingresos totales (normalizado entre 0 y 1).
Categorías de Salida:

`Riesgo Bajo`: [1, 0, 0]
`Riesgo Medio`: [0, 1, 0]
`Riesgo Alto`: [0, 0, 1]

0.4	0.4	0.8	[0, 0, 1]
0.8	0.9	0.3	[1, 0, 0]
0.5	0.7	0.6	[0, 1, 0]
0.3	0.5	0.9	[0, 0, 1]
Actividades
Implementar una red neuronal multicapa para clasificar los clientes según su riesgo.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


# Original data
original_data = np.array([
    [0.9, 0.8, 0.2],
    [0.7, 0.6, 0.5],
    [0.4, 0.4, 0.8],
    [0.8, 0.9, 0.3],
    [0.5, 0.7, 0.6],
    [0.3, 0.5, 0.9]
])

original_labels = np.array([
    [1, 0, 0],
    [0, 1, 0],
    [0, 0, 1],
    [1, 0, 0],
    [0, 1, 0],
    [0, 0, 1]
])

# Declare seed
np.random.seed(42)

# STEP 1: Generate artificial data
num_samples = 4994
data = []
labels = []

for _ in range(num_samples):
    # Generate random data
    historial_pagos = np.random.uniform(0.1, 0.9)
    ingresos_mensuales = np.random.uniform(0.1, 0.9)
    relacion_deuda_ingreso = np.random.uniform(0.1, 0.9)
    
    # To clasify what is de label to the data
    if historial_pagos > 0.7 and ingresos_mensuales > 0.7 and relacion_deuda_ingreso < 0.5:
        label = [1, 0, 0]  # Low Risk
    elif historial_pagos > 0.5 and ingresos_mensuales > 0.5 and relacion_deuda_ingreso < 0.7:
        label = [0, 1, 0]  # Medium Risk
    else:
        label = [0, 0, 1]  # Hight Risk
    
    data.append([historial_pagos, ingresos_mensuales, relacion_deuda_ingreso])
    labels.append(label)
    
artificial_data, artificial_labels =  np.array(data), np.array(labels)

# STEP 2: Mix random data when the origin balannced data
data_set = np.vstack((original_data, artificial_data))
labels = np.vstack((original_labels, artificial_labels))

# Dara sumary
print(f"Tamaño total del conjunto de datos: {len(data_set)}")
print(f"Forma del conjunto de datos: {data_set.shape}")
print(f"Forma de las etiquetas: {labels.shape}")
print("\nPrimeras 10 muestras del conjunto de datos combinado:")
for i in range(10):
    print(f"Datos: {data_set[i]}, Etiqueta: {labels[i]}")

# STEP 3: Dataset train and test split
data_set_train, data_set_test, labels_train, labels_test = train_test_split(data_set, labels, test_size=0.2, random_state=42)

# STEP 4: Create neulal model with relu
model = Sequential([
    Dense(8, input_dim=3, activation='relu'),
    Dense(6, activation='relu'),
    Dense(3, activation='softmax')
])

# STEP 5: Compile and train model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
history = model.fit(data_set_train, labels_train, epochs=200, batch_size=2, verbose=1, validation_split=0.2)

# STEP 6: Evaluar el modelo en el conjunto de prueba
loss, accuracy = model.evaluate(data_set_test, labels_test, verbose=0)
print(f"\nPrecisión en el conjunto de prueba: {accuracy:.2f}")

# Visualizar la frontera de decisión
def plot_decision_boundary(data_set, labels, model):
    x_min, x_max = data_set[:, 0].min() - 0.1, data_set[:, 0].max() + 0.1
    y_min, y_max = data_set[:, 1].min() - 0.1, data_set[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                         np.linspace(y_min, y_max, 100))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel(), np.ones_like(xx.ravel())])
    Z = np.argmax(Z, axis=1).reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.RdYlBu)
    plt.scatter(data_set[:, 0], data_set[:, 1], c=np.argmax(labels, axis=1), cmap=plt.cm.RdYlBu, edgecolor='black')
    plt.xlabel('Historial de pagos')
    plt.ylabel('Ingresos mensuales')
    plt.title('Frontera de decisión')
    plt.show()

plot_decision_boundary(data_set, labels, model)

# Graficar la pérdida y precisión durante el entrenamiento
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Pérdida de entrenamiento')
plt.plot(history.history['val_loss'], label='Pérdida de validación')
plt.title('Pérdida del modelo')
plt.xlabel('Época')
plt.ylabel('Pérdida')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Precisión de entrenamiento')
plt.plot(history.history['val_accuracy'], label='Precisión de validación')
plt.title('Precisión del modelo')
plt.xlabel('Época')
plt.ylabel('Precisión')
plt.legend()

plt.tight_layout()
plt.show()

# Probar con nuevos datos
nuevos_datos = np.array([[0.9, 0.8, 0.2], [0.5, 0.6, 0.7], [0.2, 0.3, 0.9]])
predicciones = model.predict(nuevos_datos)
for i, pred in enumerate(predicciones):
    print(f"Predicción para {nuevos_datos[i]}: {pred}")
    print(f"Clase predicha: {['Riesgo Bajo', 'Riesgo Medio', 'Riesgo Alto'][np.argmax(pred)]}\n")
```
